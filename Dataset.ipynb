```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the data
df = pd.read_csv("/content/GivenDS.csv")
```

```python
#dropping customer id
df = df.drop(['Customer ID'], axis=1)
```

```python
df
```

```python
df['Gender'].value_counts()
```

```python
df['Category'].value_counts()
```

```python
 #analysing category of item purchase with the purchase amount
 df1 = pd.DataFrame(df.groupby('Category')['Purchase Amount (USD)'].mean()).reset_index()
```

```python
df1
```

```python

df1['Purchase Amount (USD)'].plot(kind='hist', bins=20, title='Purchase Amount (USD)')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.tight_layout()
```

```python
#Analysing AGE w.r.t purchase amount
```

```python
fig, axes = plt.subplots(1,3, figsize = (15,5))

sns.kdeplot(x = 'Age', data = df, ax = axes[0])
sns.boxplot(x = 'Age', data = df, ax = axes[1])
sns.scatterplot(x = 'Age', y = 'Purchase Amount (USD)', data = df, ax = axes[2])

```

```python
#analysing through heat maps
corr = df.corr()
sns.heatmap(corr, annot =True, linewidths = 2.5 , cmap = 'coolwarm')
```

```python
 df['Size'].value_counts()
```

```python
ax = sns.countplot(x="Location", data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

```python
sns.catplot(x='Purchase Amount (USD)', y='Season', data=df, kind='bar')

# Show plot
plt.show()
```

```python
sns.catplot(x='Size', data=df, kind='count')
plt.show()
```

```python
df['Color'].value_counts()
```

```python
sns.catplot(y='Color', data=df, kind='count')
```

```python
color = pd.DataFrame(df.groupby('Color')['Purchase Amount (USD)'].mean()).reset_index().sort_values(by = 'Purchase Amount (USD)')
```

```python
sns.barplot(x = 'Purchase Amount (USD)', y = 'Color', data = color, palette = 'Set1')
plt.show()
```

```python
grp = pd.DataFrame(df.groupby('Shipping Type')['Purchase Amount (USD)'].mean()).reset_index().sort_values(by = 'Purchase Amount (USD)')
```

```python
sns.barplot(x = 'Purchase Amount (USD)', y = 'Shipping Type', data = grp, palette = 'Set1')
```
